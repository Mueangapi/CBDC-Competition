```python
import pydeck as pdk
import pandas as pd
import math

SCATTERPLOT_LAYER_DATA = "https://raw.githubusercontent.com/visgl/deck.gl-data/master/website/bart-stations.json"
df2 = pd.read_json(SCATTERPLOT_LAYER_DATA)

# Use pandas to calculate additional data
df2["exits_radius"] = df2["exits"].apply(lambda exits_count: math.sqrt(exits_count))

# Define a layer to display on a map
layer = pdk.Layer(
    "ScatterplotLayer",
    df2,
    pickable=True,
    opacity=0.8,
    stroked=True,
    filled=True,
    radius_scale=6,
    radius_min_pixels=1,
    radius_max_pixels=100,
    line_width_min_pixels=1,
    get_position="coordinates",
    get_radius="exits_radius",
    get_fill_color=[255, 140, 0],
    get_line_color=[0, 0, 0],
)
df2["latitude"] = df2["coordinates"].apply(lambda coords: coords[1])
df2["longitude"] = df2["coordinates"].apply(lambda coords: coords[0])
df2= df2[df2['longitude']<-122.34]

# Set the viewport location
view_state = pdk.ViewState(latitude=37.7749295, longitude=-122.4194155, zoom=10, bearing=0, pitch=0)

# Render
r = pdk.Deck(layers=[layer], initial_view_state=view_state, tooltip={"text": "{name}\n{address}"})
r.to_html("scatterplot_layer.html")
```

```python


#GREAT_CIRCLE_LAYER_DATA = "https://raw.githubusercontent.com/visgl/deck.gl-data/master/website/flights.json"
df = pd.read_json('th.json')

# Use pandas to prepare data for tooltip
df["from_name"] = df["from"].apply(lambda f: f["name"])
df["to_name"] = df["to"].apply(lambda t: t["name"])
df['value']=np.random.randint(10000, 100100, size=len(df))
df['radius'] = df['value'].apply(lambda exits_count: math.sqrt(exits_count))
for col in df["to"].iloc[0].keys():
    df['location'] = df["to"].apply(lambda x: x.get('coordinates'))
# Define a layer to display on a map
layer = pdk.Layer(
    "ArcLayer",
    df,
    pickable=True,
    get_stroke_width=12,
    get_source_position="from.coordinates",
    get_target_position="to.coordinates",
    get_source_color=[64, 255, 0],
    get_target_color=[0, 128, 200],
    auto_highlight=True,
)
layer2 = pdk.Layer(
    "ScatterplotLayer",
    df,
    pickable=True,
    opacity=0.8,
    stroked=True,
    filled=True,
    radius_scale=100,
    radius_min_pixels=1,
    radius_max_pixels=100,
    line_width_min_pixels=1,
    get_position="location",
    get_radius="radius",
    get_fill_color=[135,206,250],
    get_line_color=[0, 0, 0],
)

# Set the viewport location
view_state = pdk.ViewState(latitude=18.7833, longitude=98.9667, zoom=4, bearing=0, pitch=0)

# Render
r = pdk.Deck(layers=[layer,layer2], initial_view_state=view_state, tooltip={"text": "{from_name} to {to_name}"})
r.picking_radius = 10

r.to_html("macro.html")
```
